# LSTM Example 3: predict the next word in a sentence
## Notes

Below is a coding example using an LSTM model in TensorFlow to work with a sequence of strings. The example demonstrates how to process a list of strings and train an LSTM model to predict the next word in a sentence. This is a common task in natural language processing called text generation.

````python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
````

- NumPy (np) is used for numerical operations.
- TensorFlow (tf) is a deep learning library, specifically its Keras API is used here.
- Tokenizer and pad_sequences are used for text preprocessing.
- Sequential, Embedding, LSTM, and Dense are layers used to build the neural network model.

````python
# In this example, we will use a simple list of sentences to train an LSTM model to predict the next word based on the previous words in the sequence.
# Sample list of sentences (input data)
# max seq length = 5 words
sentences = [
    "Lynch believes investing in simple companies creates long term value for patient investors",
 "Lynch studies earnings growth carefully before making any investment decision in the market",
 "Lynch prefers companies with strong fundamentals and consistent profit performance over time",
 "Lynch avoids businesses with excessive debt and weak financial stability in uncertain markets",
 "Lynch focuses on understanding the business model before purchasing shares of any company",
 "Lynch believes long term growth drives stock prices higher in competitive industries",
 "Lynch emphasizes discipline research and patience in successful stock investing",
 "Lynch looks for companies with competitive advantages and sustainable earnings growth",
 "Lynch invests in businesses he understands with clear products and steady demand",
 "Lynch believes strong management teams create lasting shareholder value over time",
 "Lynch studies financial statements to evaluate company strength and growth potential",
 "Lynch prefers steady growth companies instead of speculative short term opportunities",
 "Lynch believes market volatility creates attractive buying opportunities for informed investors",
 "Lynch focuses on real earnings performance rather than temporary market trends",
 "Lynch recommends investors learn about industries before selecting individual stocks",
 "Lynch believes knowledge and research reduce long term investment risk",
 "Lynch invests patiently while allowing company performance to compound over years",
 "Lynch avoids emotional decisions during periods of market uncertainty and fear",
 "Lynch believes strong companies survive difficult economic environments successfully",
 "Lynch studies business fundamentals to identify companies with sustainable growth"
]
````

````python
# Tokenize the sentences
tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)
total_words = len(tokenizer.word_index) + 1
print(total_words)
````

- Tokenizer is used to convert the text into a numerical format (tokens).
- fit_on_texts builds a word index based on the sentences.
- total_words is the total number of unique words plus one (to account for zero indexing). 

````python
input_sequences = tokenizer.texts_to_sequences(sentences)
# Print the word index to see the mapping of words to their token indices
print("Word Index:", tokenizer.word_index)
print("Tokenize Results: ", sentences[0], input_sequences[0])
print("Tokenize Results: ", sentences[3], input_sequences[3])
````

````python
# Create input sequences and corresponding labels
input_sequences = []
for sentence in sentences:
    token_list = tokenizer.texts_to_sequences([sentence])[0]
    for i in range(1, len(token_list)):
        n_gram_sequence = token_list[:i + 1]
        input_sequences.append(n_gram_sequence)
````

- Each sentence is converted into a sequence of tokens.
- N-gram sequences are generated from each sentence to create training data. For example, from "Kakashi is reading a fiction", sequences like [kakashi], [kakashi, is], [kakashi, is, reading], etc., are created.  

````python
# Pad sequences to ensure uniform length
max_sequence_len = max(len(seq) for seq in input_sequences)
input_sequences = pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre')
````

````python
print(f' shape of input_sequence: {input_sequences.shape}')
````

- Sequences are padded to ensure they all have the same length. Padding adds zeros at the beginning of sequences shorter than the maximum length. input_sequences.shape = (21, 5)

````python
sentence = "Lynch believes investing in simple companies creates long term value for patient investors"

words = sentence.split()

for i in range(1, len(words) + 1):
    prefix = " ".join(words[:i])

    # prefix -> token ids
    seq = tokenizer.texts_to_sequences([prefix])[0]

    # pad to max_sequence_len
    padded = pad_sequences([seq], maxlen=max_sequence_len, padding='pre')[0]

    print(f"{prefix:<80} -> {padded}")
````

````python
# Split input sequences into features and labels
X = input_sequences[:, :-1]
y = input_sequences[:, -1]

# Convert labels to one-hot encoded format
y = tf.keras.utils.to_categorical(y, num_classes=total_words)
````

- The input sequences (X) consist of all but the last token of each padded sequence.
- The label (y) is the last token of each sequence, representing the word the model needs to predict.
- y is one-hot encoded to represent the output labels as categorical data. total_words = 19

````python
# Build the LSTM model
model = Sequential()
model.add(Embedding(total_words, 64))
model.add(LSTM(50, activation='relu'))
model.add(Dense(total_words, activation='softmax'))
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model: The model is trained on the input sequences (X) and labels (y) for 100 epochs.
history = model.fit(X, y, epochs=100, verbose=1)
````

- A sequential model is created.
- An Embedding layer is used to convert word indices into dense vectors of fixed size (64 dimensions).
- An LSTM layer with 50 units processes the sequences to learn patterns.
- A Dense layer with softmax activation is used for multi-class classification, outputting a probability distribution over all possible words.
- The model is compiled using the adam optimizer and categorical_crossentropy loss, suitable for multi-class classification.

````python
# Function to generate text based on seed text
def generate_text(seed_text, next_words, max_sequence_len):
    for _ in range(next_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        # print(token_list)
        token_list = pad_sequences([token_list], maxlen=max_sequence_len - 1, padding='pre')
        # print(token_list)
        predicted = np.argmax(model.predict(token_list, verbose=0), axis=-1)
        # print(predicted)
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        seed_text += " " + output_word
    return seed_text
````

```def generate_text()``` function generates text by predicting the next word based on the seed text.
- It iteratively predicts the next word, appends it to the seed text, and continues for the specified number of words (next_words).
- The text generation function is called with a seed text (e.g., "Naruto and Sasuke") and generates the next K words.

````python
# Test the text generation - Test 3
seed_text = "Lynch believes in"
next_words = 5
generated_text = generate_text(seed_text, next_words, max_sequence_len)
print(f"Generated Text: {generated_text}")
````
